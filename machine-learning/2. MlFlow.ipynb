# MlFlow Turtorials

## Lifecycle of a Data Science project

Data engineer will aggregiate data using ETL Pipeline from various sources such as API's or streaming data's and push to a databas.

Data Scientist leverage MlFlow for:
1. Experiment tracking and hypothesis testing.
2. Code structuring.
3. Model packaging and dependancy management.
4. Evaluating hyper parameter tuning.
    - Train Model -> Parameters
    - Tracks Hyper parameter tuning
    -- Visualisation graphs
5. Compare the results of models retraining over time. 

MLOps Professionals
1. Manage the lifecycle of trained models, both pre and post deployment.
2. Deploy models securely to the production enviornment. 
3. Manage entire deployment dependencies. 

Prompt Engineering Users
1. Evaluate and experiment with large language models.
2. Create custom prompts and experiment with them. 
3. Deciding on the best base model suitable for your usecase. 


Usecases of MlFlow
1. Experiment Tracking -> Track parameters and Metrics via UI.
2. Model selection and deployment. 
3. Model performance monitoring. 
4. Collaborative in project.
